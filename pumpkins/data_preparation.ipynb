# Calculate Nearest Station to Each Pumpkin

```python
def distance(lat_a, long_a, lat_b, long_b):
    return ((float(lat_b) - float(lat_a))**2 + (float(long_b) - float(long_a))**2)**.5
```

```python
us_stations.head(5)
```

```python
pumpkin_clean.head(5)
```

```python
pumpkin_clean["nearest_station"] = ""
for i in range(len(pumpkin_clean)):
    
    state_stations = us_stations[us_stations.state == pumpkin_clean.loc[i,"state_code"]]
    
    pumpkin_lat = pumpkin_clean.loc[i,"latitude"]
    pumpkin_long = pumpkin_clean.loc[i,"longitude"]
    
    closest_distance = 1000000000000000000.0
    nearest_station = ""
    
    for index, row in state_stations.iterrows():
        check = distance(pumpkin_lat, pumpkin_long, row["latitude"], row["longitude"])
        if check < closest_distance:
            closest_distance = check
            nearest_station = row["id"]
    
    pumpkin_clean.loc[i, "nearest_station"] = nearest_station
    print(pumpkin_clean.loc[i, "nearest_station"])
    print(closest_distance)
    print(f"{i+1} pumpkins located. {len(pumpkin_clean) - (i+1)} left. {(i+1)/len(pumpkin_clean)*100:.0f}% done.\n")
```
