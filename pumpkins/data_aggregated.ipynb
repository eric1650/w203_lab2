# Create Regression Features from Weather Data

```python
import pandas as pd
import datetime as dt
```

```python
weather = pd.read_csv("weather.csv")
```

```python
weather
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather
```

```python
weather['year'] = weather['date'].dt.year
weather
```

```python
weather.drop(['measurement_flag', 'quality_flag', 'source_flag', 'time_of_observation'], axis=1, inplace=True)
weather
```

```python
rainfall = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).sum().reset_index()
rainfall['avg'] = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).agg({'value': ['mean'],}).values
rainfall['stdev'] = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).agg({'value': ['std'],}).values
rainfall.columns = ['station_id','year','element','rain_total','rain_daily_avg','rain_daily_stdev']
rainfall.drop(["element"], axis=1, inplace=True)
rainfall
```

```python
rainfall.isna().sum()
```

```python
temp = weather[weather.element.isin(['TMIN','TMAX'])]
temp.element.value_counts()
```

```python
temp_min = temp[temp.element == "TMIN"]
temp_min.columns = ['station_id', 'date', 'element', 'TMIN', 'year']
temp_min_avg = temp_min.drop(["date","element"], axis=1).groupby(["station_id","year"]).mean().reset_index()

temp_max = temp[temp.element == "TMAX"]
temp_max.columns = ['station_id', 'date', 'element', 'TMAX', 'year']
temp_max_avg = temp_max.drop(["date","element"], axis=1).groupby(["station_id","year"]).mean().reset_index()



# temp_avg = temp[temp.element == "TMIN"].groupby(['station_id', 'year', 'element']).mean().reset_index()
# temp_avg['TMAX_avg'] = temp[temp.element == "TMAX"].groupby(['station_id', 'year', 'element']).mean().values
# temp_avg.columns = ['station_id','year','element','TMIN_AVG','TMAX_AVG']
# temp_avg.drop(['element'], axis=1, inplace = True)
# temp_avg
```

```python
print(len(temp_min_avg))
print(len(temp_max_avg))

print(temp_min_avg.columns)
print(temp_max_avg.columns)
```

```python
temp_avg = temp_min_avg.merge(temp_max_avg, how="inner", left_on=['station_id', 'year'], right_on=['station_id', 'year'])
temp_avg
```

```python
#consolidating weather data into one table - including temp, rain
weather_consol=rainfall.merge(temp_avg, how="inner", left_on=["station_id","year"], right_on=["station_id","year"])
weather_consol
```

```python
weather_consol.isna().sum()
```

# Combine Weather and Pumpkin Data

```python
pumpkin = pd.read_csv("pumpkin.csv")
pumpkin
```

```python
pumpkin_consol = pumpkin.merge(weather_consol, how="inner", left_on=["nearest_station","year"], right_on=["station_id","year"])
pumpkin_consol
```
