# Create Regression Features from Weather Data

```python
import pandas as pd
import datetime as dt
```

```python
weather = pd.read_csv("weather.csv.zip")
```

```python
weather
```

```python
weather['date'] = pd.to_datetime(weather['date'])
weather
```

```python
weather['year'] = weather['date'].dt.year
weather
```

```python
weather.drop(['measurement_flag', 'quality_flag', 'source_flag', 'time_of_observation'], axis=1, inplace=True)
weather
```

```python
rainfall = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).sum().reset_index()
rainfall['avg'] = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).agg({'value': ['mean'],}).values
rainfall['stdev'] = weather[weather.element == 'PRCP'].groupby(['station_id', 'year', 'element']).agg({'value': ['std'],}).values
rainfall.columns = ['station_id','year','element','rain_total','rain_daily_avg','rain_daily_stdev']
rainfall
```

```python
temp = weather[weather.element.isin(['TMIN','TMAX'])]
temp.element.value_counts()
```

```python
temp_avg = temp[temp.element == "TMIN"].groupby(['station_id', 'year', 'element']).mean().reset_index()
temp_avg['TMAX_avg'] = temp[temp.element == "TMAX"].groupby(['station_id', 'year', 'element']).mean().values
temp_avg.columns = ['station_id','year','element','TMIN_AVG','TMAX_AVG']
temp_avg.drop(['element'], axis=1, inplace = True)
temp_avg
```
