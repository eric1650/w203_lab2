```python
us_pumpkins["city_state"] = us_pumpkins["city"].str.title() + ", " + us_pumpkins["state_prov"]
us_pumpkins.head()
```

```python
us_cities_kaggle = us_cities_kaggle.merge(state_abbrev, how="left", left_on="Region", right_on="code")
us_cities = us_cities_kaggle.drop(["Country", "City", "Region", "Population"], axis=1)
us_cities.head(10)
```

```python
us_cities.state = us_cities.state.str.title()
us_cities.head()
```

```python
us_cities["city_state"] = us_cities["AccentCity"] + ", " + us_cities["state"]
us_cities.head()
```

```python
count = 0
absent_list = []

for city in us_pumpkins.city_state.values:
    if city in set(us_cities_kaggle.city_state.values):
        count += 1
    else:
        absent_list.append(city)

print(count, "out of", len(us_pumpkins))
```

```python
us_pumpkins_latlong = us_pumpkins.merge(us_cities, how="left", left_on="city_state", right_on="city_state")
us_pumpkins_latlong.head()
```

```python
us_pumpkins_filtered = us_pumpkins_latlong.dropna(subset=['Latitude', 'Longitude'])
us_pumpkins_filtered.index = range(len(us_pumpkins_filtered))
print(f"{len(us_pumpkins) - len(us_pumpkins_filtered)} pumpkins omitted due to failure to match city")
```

```python
us_pumpkins_filtered.head(10)
```

```python
def distance(lat_a, long_a, lat_b, long_b):
    return ((float(lat_b) - float(lat_a))**2 + (float(long_b) - float(long_a))**2)**.5
```

```python
us_pumpkins_filtered["Nearest_Station"] = ""
count=0
for i in range(len(us_pumpkins_filtered)):
    state_stations = us_stations[us_stations.state == us_pumpkins_filtered.loc[i,"code"]]
    pumpkin_lat = us_pumpkins_filtered.loc[i,"Latitude"]
    print(pumpkin_lat)
    pumpkin_long = us_pumpkins_filtered.loc[i,"Longitude"]
    print(pumpkin_long)
    closest_distance = 1000000000000000000.0
    Nearest_Station = ""
    for index, row in state_stations.iterrows():
        check = distance(pumpkin_lat, pumpkin_long, row["latitude"], row["longitude"])
        if check < closest_distance:
            closest_distance = check
            Nearest_Station = row["id"]
    us_pumpkins_filtered.loc[i, "Nearest_Station"] = Nearest_Station
    print(Nearest_Station)
    print(closest_distance)
    count+=1
    print(count)
    
us_pumpkins_filtered.head(10)
```
